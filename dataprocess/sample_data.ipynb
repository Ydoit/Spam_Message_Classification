```python
import csv
sample_data = {}
with open ('../data/sample1.csv','r') as f1,open('../data/sample2.csv','r') as f2,open('../data/sample3.csv','r') as f3:
    reader1 = csv.reader(f1)
    reader2 = csv.reader(f2)
    reader3 = csv.reader(f3)
    next(reader1)
    next(reader2)
    next(reader3)
    for row in reader1:
        sample_data[row[1]] = row[0]
    for row in reader2:
        sample_data[row[1]] = row[0]
    for row in reader3:
        sample_data[row[1]] = row[0]
test_data={}
test_data1={}
test_data2={}
test_data3={}
with open ('../rawdata/spam.csv','r') as f1,open('../rawdata/sms_pub1.csv','r') as f2 ,open('../rawdata/sms_pub2.csv','r') as f3:
    reader1 = csv.reader(f1)
    reader2 = csv.reader(f2)
    reader3 = csv.reader(f3)
    next(reader1)
    next(reader2)
    next(reader3)
    for row in reader1:
        if row[1] not in sample_data.keys():
            test_data[row[1]] = row[0]
            test_data1[row[1]] = row[0]
    for row in reader2:
        
        if row[1] not in sample_data.keys():
            test_data[row[1]] = row[0]
            test_data2[row[1]]=row[0]
    for row in reader3:
        
        if row[1] not in sample_data.keys():
            test_data[row[1]] = row[0] 
            test_data3[row[1]]=row[0]
```

```python
import random 
test_part=random.sample(list(test_data.keys()),40000)
```

```python
with open('../data/test1.csv','w') as f1,open('../data/test2.csv','w') as f2,open('../data/test3.csv','w') as f3:
    writer1 = csv.writer(f1)
    writer2 = csv.writer(f2)
    writer3 = csv.writer(f3)
    writer1.writerow(['label','content'])
    writer2.writerow(['label','content'])
    writer3.writerow(['label','content'])
    for key in test_data:
        if key in test_data1.keys():
            writer1.writerow([test_data1[key],key])
        if key in test_data2.keys():
            writer2.writerow([test_data2[key],key])
        if key in test_data3.keys():
            writer3.writerow([test_data3[key],key])
```

```python
with open('../data/test.csv','w') as f:
    writer = csv.writer(f)
    writer.writerow(['label','content'])
    for key in test_part:
        writer.writerow([test_data[key],key])
```
